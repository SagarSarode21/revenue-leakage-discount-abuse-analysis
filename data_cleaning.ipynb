```python
import openpyxl
import pandas as pd       # For data handling
import numpy as np        # For numeric calculations

```

```python


file_path = r"C:\Users\Administrator\Downloads\online_retail_II.xlsx"

# Load both sheets
df1 = pd.read_excel(file_path, sheet_name='Year 2009-2010')
df2 = pd.read_excel(file_path, sheet_name='Year 2010-2011')

# Combine into single dataframe
df = pd.concat([df1, df2], ignore_index=True )
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Check first 5 rows
df.head()
```

```python
# Remove zero quantity rows (possible returns)
df = df[df['Quantity'] != 0]

# Drop duplicates
df.drop_duplicates(inplace=True)

# Handle missing CustomerID or Description
df = df.dropna(subset=['Customer ID', 'Description'])
df.head()
```

```python
df['TransactionID'] = df['Invoice'].astype(str) + '_' + df['StockCode'].astype(str)
df.head()
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
# Random discount between 0-30% of revenue
df['Discount'] = df['Revenue'] * np.random.uniform(0, 0.3, size=len(df))
```

```python
# Assume 20% margin (or adjust per category)
df['Cost'] = df['Revenue'] * 0.8
df['Profit'] = df['Revenue'] - df['Cost'] - df['Discount']
```

```python
# Flag invoices starting with 'C' as returns
df['ReturnFlag'] = df['Invoice'].str.startswith('C').astype('Int64')
df['RefundAmount'] = df['ReturnFlag'] * df['Revenue']
df.head()

```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df.head()
```

```python
df.to_csv("OnlineRetail_Cleaned.csv", index=False)
df.head()
```
